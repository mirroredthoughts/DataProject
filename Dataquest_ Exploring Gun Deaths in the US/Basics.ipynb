# US Gun Deaths Guided Project Solutions


```python
import csv
import datetime
data=list(csv.reader(open("guns.csv",'r')))
print(data[:5])
```

```python
headers=data[0]
data=data[1:]
print(headers)
print(data[:5])
```

```python
years = [r[1] for r in data]
print(years)
```

# Counting Gun Deaths By Year

```python
year_counts={}
for y in years:
    if y not in year_counts:
        year_counts[y]=1
    else:
        year_counts[y]+=1
        
print(year_counts)
```

```python
dates=[datetime.datetime(year=int(r[1]),month=int(r[2]),day=1) for r in data]
print(dates[:5])



```

```python
date_counts={}
for d in dates:
    if d not in date_counts:
        date_counts[d]=1
    else:
        date_counts[d]+=1
print(date_counts)
    
```

# Exploring Gun Deaths By Race And Sex

```python
sex_counts={}
for s in data:
    sex=s[5]
    if sex not in sex_counts:
        sex_counts[sex]=1
    else:
        sex_counts[sex]+=1
print(sex_counts)

race_counts={}
for r in data:
    race=r[7]
    if race not in race_counts:
        race_counts[race]=1
    else:
        race_counts[race]+=1
print(race_counts)

```

```python
census= list(csv.reader(open("census.csv",'r')))
print(census)
```

# Computing Rates Of Gun Deaths Per Race

```python
mapping={}

mapping['Asian/Pacific Islander']=int(census[1][14])+int(census[1][15])
mapping['Black']=census[1][12]
mapping['Hispanic']=census[1][11]
mapping['Native American/Native Alaskan']=census[1][13]
mapping['White']=census[1][10]

race_per_hundredk={}

for k,v in race_counts.items():
    race_per_hundredk[k]=(int(v)/int(mapping[k]))*100000
    
print(race_per_hundredk)    
        
    


```

```python
intents=[row[3] for row in data]
races = [row[7] for row in data]

homicide_race_counts={}

for i,race in enumerate(races):
    if race not in homicide_race_counts:
        homicide_race_counts[race] = 0
    if intents[i]=="Homicide":
        homicide_race_counts[race]+=1
        
race_per_hundredk={}

for k,v in homicide_race_counts.items():
    race_per_hundredk[k]=int(v)/int(mapping[k])*100000
    
race_per_hundredk
```
